```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Data1-A.csv")
df.head(9)
```

```python
mean = df.drop(['All_India GDP','West Bengal1','Andaman & Nicobar Islands','Chandigarh','Delhi','Puducherry'],axis=1).iloc[6:9].mean(axis=0)
mean = mean.sort_values()
mean
```

```python
p = mean.plot(figsize=(15,5),kind="bar",fontsize=16)
p.set_title("Average Growth of States", fontsize=18);
p.set_xlabel("States", fontsize=18);
p.set_ylabel("Average", fontsize=18);



```
