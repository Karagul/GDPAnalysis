```python
state_name=['Andhra_Pradesh','Arunachal_Pradesh','Assam','Bihar','Chhattisgarh','Gujarat','Haryana','Himachal_Pradesh','Jharkhand','Karnataka','Kerala','Madhya_Pradesh','Maharashtra','Manipur','Meghalaya','Mizoram','Nagaland','Odisha','Punjab','Rajasthan','Sikkim','Tamil_Nadu','Telangana','Tripura','Uttar_Pradesh','Uttarakhand']
#state_name=['Andhra_Pradesh','Arunachal_Pradesh','Assam','Bihar','Chhattisgarh','Gujarat','Haryana','Himachal_Pradesh','Jharkhand','Manipur','Meghalaya','Mizoram','Nagaland','Odisha','Karnataka','Kerala','Madhya_Pradesh','Maharashtra','Punjab','Rajasthan','Sikkim','Tamil_Nadu','Telangana','Tripura','Uttar_Pradesh','Uttarakhand']
```

```python
dfs=[]
for state in state_name:
    dfs.append(pd.read_csv('{}.csv'.format(state),encoding='ISO-8859-1'))
    
#merging all the csv of diffrent states
```

```python
dfs
```

```python
for i in range(len(dfs)):
    dfs[i]=dfs[i].loc[:,['Item','2014-15']]
    dfs[i]['State']=state_name[i]
```

```python
columns=['TOTAL GSVA at basic prices','Taxes on Products','Subsidies on products','Gross State Domestic Product',"Population ('00)",'Per Capita GSDP (Rs.)','Agriculture, forestry and fishing','Crops','Livestock','Forestry and logging','Fishing and aquaculture','Mining and quarrying','Primary','Manufacturing','Electricity, gas, water supply & other utility services','Construction','Secondary','Trade, repair, hotels and restaurants','Hotels & restaurants','Transport, storage, communication & services related to broadcasting','Financial services','Real estate, ownership of dwelling & professional services','Public administration','Other services','Communication & services related to broadcasting','Storage','Air transport','Water transport','Tertiary']
```

```python
data=[ [] for _ in range(len(columns))]
for r in range(len(columns)):
    for i in range(len(dfs)):
        data[r].append(dfs[i].loc[(dfs[i].Item==columns[r]),'2014-15'].astype(int))
        
dictionary={}
dictionary['State']=state_name
dictionary.update(dict(zip(columns,data)))
gsdp=pd.DataFrame(dictionary)
gsdp
```

```python
my_cols=['TOTAL GSVA at basic prices','Taxes on Products','Subsidies on products','Gross State Domestic Product',"Population ('00)",'Per Capita GSDP (Rs.)','Agriculture, forestry and fishing','Crops','Livestock','Forestry and logging','Fishing and aquaculture','Mining and quarrying','Primary','Manufacturing','Electricity, gas, water supply & other utility services','Construction','Secondary','Trade, repair, hotels and restaurants','Hotels & restaurants','Transport, storage, communication & services related to broadcasting','Financial services','Real estate, ownership of dwelling & professional services','Public administration','Other services','Communication & services related to broadcasting','Storage','Air transport','Water transport','Tertiary']
```

```python
for cols in my_cols:
    gsdp[cols]=gsdp[cols].apply(lambda x:x.iloc[0])
    
```

```python
gsdp=gsdp.sort_values('Per Capita GSDP (Rs.)')
gsdp['Primary']
#till here we have sort out the dataframe(gsdp) according to our need.
```

```python
bar=gsdp.plot.bar(x='State',y='Per Capita GSDP (Rs.)',figsize=(15,5),fontsize=16)
bar.set_title("GDP per capita(2014-15)", fontsize=18);
bar.set_xlabel("States", fontsize=18);
bar.set_ylabel("Per Capita GSDP(Rs)", fontsize=18);

#Plot the GDP per capita for all the states.
```

```python
#Identify the top-5 and the bottom-5 states based on GDP per capita
#top5 are : Sikkim,Haryana,Kerala,Uttarakhnad,Maharashtra
#bottom5 are: Bihar,Uttar_Pradesh,Manipur,Assam,Jharkhand.
```

```python
#Find the ratio of highest per capita GDP to the lowest per capita GDP.
highest=gsdp['Per Capita GSDP (Rs.)'].max()
lowest=gsdp['Per Capita GSDP (Rs.)'].min()
ratio=highest/lowest
ratio
#ratio of highest per capita GDP to the lowest per capita GDP = 7.07(aprrox)
```

```python
gsdp['primary%']=gsdp['Primary']/gsdp['Gross State Domestic Product']*100
gsdp['secondary%']=gsdp['Secondary']/gsdp['Gross State Domestic Product']*100
gsdp['tertiary%']=gsdp['Tertiary']/gsdp['Gross State Domestic Product']*100
```

```python
polt=gsdp.plot(x="State", y=["primary%", "secondary%", "tertiary%"], kind="bar",figsize=(15,5),fontsize=16)
polt.set_title("% contibution(2014-15)", fontsize=18);
polt.set_xlabel("States", fontsize=18);
polt.set_ylabel("Percentage Contribution", fontsize=18);

#Plot the percentage contribution of primary, secondary and tertiary sectors as a percentage of total GDP for all the states.
```

```python
polt=gsdp.plot(x="State", y=["primary%", "secondary%", "tertiary%"], kind="bar",stacked =True,figsize=(15,5),fontsize=16)
polt.set_title("% contibution(2014-15)", fontsize=18);
polt.set_xlabel("States", fontsize=18);
polt.set_ylabel("Percentage Contribution", fontsize=18);

#Plot the percentage contribution of primary, secondary and tertiary sectors as a percentage of total GDP for all the states.using stacked bargraph
```
