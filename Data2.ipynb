```python
import os
os.chdir(r'C:\Users\abc\Desktop\New folder')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

```python
state_name=['Andhra_Pradesh','Arunachal_Pradesh','Assam','Bihar','Chhattisgarh','Gujarat','Haryana','Himachal_Pradesh','Jharkhand','Karnataka','Kerala','Madhya_Pradesh','Maharashtra','Manipur','Meghalaya','Mizoram','Nagaland','Odisha','Punjab','Rajasthan','Sikkim','Tamil_Nadu','Telangana','Tripura','Uttar_Pradesh','Uttarakhand']
```

```python
dfs=[]
for state in state_name:
    dfs.append(pd.read_csv('{}.csv'.format(state),encoding='ISO-8859-1'))
```

```python
for i in range(len(dfs)):
    dfs[i]=dfs[i].loc[:,['Item','2014-15']]
    dfs[i]['State']=state_name[i]
```

```python
columns=['Per Capita GSDP (Rs.)']
```

```python
data=[ [] for _ in range(len(columns))]
for r in range(len(columns)):
    for i in range(len(dfs)):
        data[r].append(dfs[i].loc[(dfs[i].Item==columns[r]),'2014-15'].astype(int))
        
dictionary={}
dictionary['State']=state_name
dictionary.update(dict(zip(columns,data)))
gsdp=pd.DataFrame(dictionary)
gsdp
```

```python
my_cols=["Per Capita GSDP (Rs.)"]
for cols in my_cols:
    gsdp[cols]=gsdp[cols].apply(lambda x:x.iloc[0])
```

```python
gsdp
```

```python
dropout=pd.read_csv("dropout.csv")
```

```python
dropout=dropout.drop(['Sl. No.','Primary - 2012-2013','Primary - 2014-2015','Upper Primary - 2012-2013','Upper Primary - 2013-2014','Secondary - 2012-2013','Secondary - 2013-2014','Senior Secondary - 2012-2013','Senior Secondary - 2013-2014'],axis=1)
```

```python
dropout.replace(['Chhatisgarh','Uttrakhand','Andhra Pradesh','Arunachal Pradesh','Himachal Pradesh','Madhya Pradesh','Tamil Nadu'],['Chattisgarh','Uttarakhnad','Andhra_Pradesh','Arunachal_Pradesh','Himachal_Pradesh','Madhya_Pradesh','Tamil_Nadu'])
```

```python
dropout=dropout.merge(gsdp,left_on='Level of Education - State',right_on='State')
```

```python
dropout=dropout[["Level of Education - State", "Primary - 2014-2015.1", 'Upper Primary - 2014-2015', "Secondary - 2014-2015", "Senior Secondary - 2014-2015",'Per Capita GSDP (Rs.)']]
```
